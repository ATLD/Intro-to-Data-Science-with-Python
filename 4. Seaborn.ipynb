# Seaborn Intro
Seaborn is a plotting library built directly on top of matplotlib. You will sometimes hear it being called a 'wrapper' for matplotlib. It makes plotting much easier by offering functions that are very straightforward and simple. 

```python
import seaborn as sns
import pandas as pd
%matplotlib inline
```

```python
housing = pd.read_csv('data/housing.csv')
```

```python
housing['Sa']
```

```python
sns.jointplot('GrLivArea', 'SalePrice', data=housing)
```

```python
g = sns.lmplot('GrLivArea', 'SalePrice', data=housing, hue='OverallQual', col='MSZoning', col_wrap=2)
g.set(ylim=(0, 600000))
```

```python
sns.violinplot(x='OverallQual', y='GrLivArea', data=housing)
```

```python
housing['year_cat'] = pd.cut(housing.YearBuilt, 6)
housing.year_cat.value_counts()
```

```python
sns.factorplot(x='OverallQual', y='GrLivArea', data=housing, kind='violin', col='year_cat', col_wrap=3)
```
